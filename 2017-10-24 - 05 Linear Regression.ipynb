# Linear Regression

Let's fabricate some data that shows a roughly linear relationship between page speed and amount purchased:

```python
import numpy as np
from pylab import *
import matplotlib.pyplot as plt

# generate some random data
pageSpeeds = np.random.normal(3.0, 1.0, 1000)
purchaseAmount = 100 - (pageSpeeds + np.random.normal(0, 0.1, 1000)) * 3

# plot the generated data
plt.scatter(pageSpeeds, purchaseAmount)
plt.show()

# close the plt object
plt.close()
```

As we only have two features, we can keep it simple and just use scipy.state.linregress:

```python
from scipy import stats

# train a linear regression model and output its parameters
lin_model = stats.linregress(pageSpeeds, purchaseAmount)

#print(intercept, slope)
print(lin_model)
```

Not surprisngly, our R-squared value shows a really good fit:

```python
# calculate R² from r and print it
# surprisingly, R² is r squared
print(lin_model[2] ** 2)
```

Let's use the slope and intercept we got from the regression to plot predicted values vs. observed:

```python
import matplotlib.pyplot as plt

# implement model as a prediction method
def predict(x):
    return lin_model[0] * x + lin_model[1]

fitLine = predict(pageSpeeds)

# drow dots into the plot object
plt.scatter(pageSpeeds, purchaseAmount)

# draw line into the plot object
plt.plot(pageSpeeds, fitLine, c='r')
plt.show()
plt.close()
```
