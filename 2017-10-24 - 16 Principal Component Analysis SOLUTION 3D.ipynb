```python
from sklearn.decomposition import PCA
from sklearn.datasets import load_iris
import pylab as pl
from itertools import cycle

import matplotlib.pyplot as plt

iris = load_iris()
X = iris.data
```

```python
X2 = iris.data
pca2 = PCA(n_components=3, whiten=True).fit(X)
X_pca2 = pca2.transform(X)
print(pca2.components_)
```

```python
print(pca2.explained_variance_ratio_)
```

```python
colors = cycle('rgb')

# Name of the targe species
target_ids = range(len(iris.target_names))

pl.figure()


for i, c, label in zip(target_ids, colors, iris.target_names): 
    plt.scatter(X_pca2[iris.target == i, 0], X_pca2[iris.target == i, 1], X_pca2[iris.target == i, 2], label=label)

pl.xlabel("Principal Component 1")
pl.ylabel("Principal Component 1")
pl.title("Iris Data reduced to 1 PC")
pl.legend()
pl.show()
```
